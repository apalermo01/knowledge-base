# Adversarial example generation tutorial

source: https://pytorch.org/tutorials/beginner/fgsm_tutorial.html

One of the security concerns around ML models is the threat of attacks meant to fool the model.

2 types of attacks: white-box (attacker has full access to the model) and black-box (attacker only has access to inputs and outputs)

2 kinds of goals: misclassification (make the model predict something incorrect) and source/target misclassification (make the model predict a specific class)

In this tutorial - doing a **Fast Gradient Sign Attack (FGSM)** (https://arxiv.org/abs/1412.6572)
